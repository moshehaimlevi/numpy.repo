```python

import numpy as np
from IPython.display import display
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
file_path = '/content/drive/My Drive/sport_athletes.csv'

df = pd.read_csv(file_path)
display(df)
```

```python
######################## QUESTION 1 ########################
df['Max weight'] = df['Weight(kg)'].max()
display()
df['Max height'] = df['Height(cm)'].max()
display()
df['Max wins'] = df['Wins'].max()
display(df)
#### OPTION 2 ####
######################## QUESTION 1 ########################
df_max = df.groupby(['Height(cm)', 'Weight(kg)', 'Wins']).max()
display(df_max)
```

```python
######################## QUESTION 2 ########################
df_highest = df.nlargest(2, 'Height(cm)')
display(df_highest)


```

```python
######################## QUESTION 4 ########################
new_rows = pd.DataFrame({
    'Country': ['Canada', 'USA'],
    'Sport Type': ['Hockey', 'Skiing'],
    'Athlete Name': ['Wayne Gretzky', 'Lindsey Vonn'],
    'Height (cm)': [183, 172],
    'Weight (kg)': [84, 68],
    'Victory': [3, 4],
    'Olympic Date': ['01/02/2002', '12/02/2010']
})

df_new = pd.concat([df, new_rows])
display(df_new)
```
