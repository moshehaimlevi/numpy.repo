```python
import numpy as np
from IPython.display import display
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
file_path = '/content/drive/My Drive/mobile.csv'

df = pd.read_csv(file_path)
display(df)
```

```python
########################### Q1 ###########################
df['first name'] = df['Customer Name'].apply(lambda x: x.split()[0])
display(df)
```

```python
########################### Q2 ###########################
def get_carrier(phone_number):
    if phone_number.startswith('050-'):
        return 'Pelephone'
    elif phone_number.startswith('051-'):
        return 'Golan'
    elif phone_number.startswith('052-'):
        return 'Cellcom'
    elif phone_number.startswith('053-'):
        return 'Motorola'
    else:
        return 'Unknown'


df['Phone Number'] = df['Phone Number'].apply(get_carrier)


display(df)
```
